# univariate analysis

```python
#x_axis = data_5.columns.tolist() #Discount List
x_axis=data_2[['Sales','Quantity','Discount','Profit']]
for x in x_axis:
    fig, axes = plt.subplots(1, 2, figsize=(18,4))
    sns.distplot(data_2[x], ax=axes[0], kde = False)
    sns.boxplot(data_2[x], ax=axes[1], orient = "h", showmeans = True, color = "pink")
    fig.suptitle('Univariate Analysis of ' + x, fontsize=16)

```

```python
num_corr = data_2.corr()


mask = np.zeros_like(num_corr)

mask[np.triu_indices_from(mask)] = True

with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(18, 9))
    ax = sns.heatmap(num_corr , mask=mask, vmin = -1,vmax=1, annot = True, cmap="YlGnBu")

  
```


## <font color='blue'> Observation: </font>
    1.high corelation  seen between profit and sales column

```python
#Lets see how sub-categories are distributed wrt to category
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=data_2,color='y')
plt.show()
```

```python
plt.figure(figsize=(12,10))
data_2['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
data_2.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.show()
```

## <font color='blue'> Observation: </font>
    1.Highest profit is earned in Copiers while Selling price for Chairs and Phones is extremely high 
    2. people dont prefer to buy Tables and Bookcases from Superstore. Hence these departments are in loss




```python
data_2.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Category')
plt.show()
```

## <font color='blue'> Observation: </font>
    1. Highest profit seen in technology and office supplies category
 


```python
data_2.groupby('Segment')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Segment')
plt.show()
```

## <font color='blue'> Observation: </font>
    1.consumer and corporate segment have more profit
 


```python
data_2.groupby('Ship Mode')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Ship mode')
plt.show()
```

    1.standard class made more profit
 

```python
#Count of Sub-Category region wise
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=data_2)
plt.show()
```

```python
data_2['Cost']=data_2['Sales']-data_2['Profit']
data_2['Cost'].head()
```

```python
data_2['Profit %']=(data_2['Profit']/data_2['Cost'])*100
```

```python
#Products with high Profit Percentage 
data_2.sort_values(['Profit %','Sub-Category'],ascending=False).groupby('Profit %').head(5)
```

## <font color='blue'> Observation: </font>
    1.highest profit was gain in  sales of phones and papers
    2.Loss in sales of Appliances

```python
bar_df = pd.DataFrame(data_2.groupby(["State"])["Sales"].nunique())
bar_df_1=bar_df.sort_values(by='Sales',ascending=False)
plt.figure(figsize=(10,15))
sns.barplot(x='Sales',y = bar_df_1.index ,data = bar_df_1)
plt.xlabel('sales',fontsize=13)
plt.ylabel('States',fontsize=13)
plt.title('sales across State',fontsize=16)
plt.grid()
plt.show()
```
